#### Have a look at the speed:

```python
print('Mean speed {} km/h and variance of speed {}'.format(np.mean(v_roi), np.var(v_roi)))
print('Mean speed {} km/h and variance of speed {}'.format(np.mean(v_va), np.var(v_va)))
print('Mean speed {} km/h and variance of speed {}'.format(np.mean(v_ha), np.var(v_ha)))
```

```python
def get_indices_of_speed(v, d=0.1): # d = max deviation of mean => 10 %
    too_fast = []
    too_slow = []
    v_mean = np.mean(v)
    for i in range(len(v)):
        if v[i] > v_mean*(1+d):
            too_fast = np.append(too_fast, v[i])
        elif v[i] < v_mean*(1-d):
            too_slow = np.append(too_slow, v[i])
    print('Number of too slow crossings: {}, too fast crossings: {}'.format(len(too_slow), len(too_fast)))
    return too_slow, too_fast
```

```python
tooslow_roi, toofast_roi = get_indices_of_speed(v_roi)
tooslow_va, toofast_va = get_indices_of_speed(v_va)
tooslow_ha, toofast_ha = get_indices_of_speed(v_ha)
```
